```python
import numpy as np
import cvxpy as cv
```

```python
A = np.array([
    [0.9, 0.45],
    [-0.18, 0.9]
])

ut = [-1.5, 1.75]
for _ in range(30):
    ut = A @ ut
```

```python
np.diag([1,1,1])
np.eye(2)
```

```python
P = cv.Variable((2,2), symmetric=True)
cv.Problem(cv.Minimize(0), 
          [
              P >> np.eye(2),
              P - A.T @ P @ A >> np.eye(2)
          ]).solve()
```
