```python
import cvxpy as cp
```

```python
alpha = cp.Variable()
A = cp.bmat([
    [1, alpha, alpha],
    [alpha, 1, alpha],
    [alpha, alpha, 1]
])
```

```python
min_alpha = cp.Problem(cp.Minimize(alpha), [A >> 0]).solve()
```

```python
Q = sqrtm(
    [[1, min_alpha, min_alpha],
     [min_alpha, 1, min_alpha],
     [min_alpha, min_alpha, 1]]
)
print(min_alpha, Q, sep='\n')
```
