```python
edges = [
    (0, 1),
    (0, 2),
    (2, 3),
    (1, 3),
    (1, 4),
    (3, 4)
]
```

```python
import numpy as np
import cvxpy as cv
from scipy.linalg import sqrtm
```

```python
X = cv.Variable((5,5), symmetric=True)
constraints = [X >> 0]
constraints += [X[i,i] == 1 for i in range(5)] # only vectors with unit norm
```

```python
objective = sum ( 0.5 * (1-X[i,j]) for i,j in edges )
```

```python
prob = cv.Problem(cv.Maximize(objective), constraints)
```

```python
prob.solve()
```

```python
x = sqrtm(X.value)
print(x)
```
