```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as alg
```

```python
rng = np.random.default_rng(42)
```

```python
# creacion de datos
n, d = 100, 1
x = np.linspace(-3, 3, n)
e = rng.uniform(-1,1, n)

a, b = 1, 0
y = np.sin(a * x) + b * x + 0.2*e
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1); ax1.plot(x,y)
ax2 = fig.add_subplot(1,2,2); ax2.plot(x, np.sin(x))
```

```python
# creacion de la matriz de diseño
X = np.column_stack([np.ones(n), x])
X = np.hstack([np.ones((n,1)), x.reshape(-1,1)])
X 
```

```python
tau_default = 0.1
```

```python
def w(x, X, tau):
    a = -1 / ( 2*(tau**2) )
    return np.exp( a * (X[:, 1] - x)**2)
w(0, X)[int(n/2) -4: int(n/2) + 4]
```

```python
def theta(x, X, Y, tau):
    # X and Y are x-train y-train data sample
    # theta such that solves X.T W X theta = X.T W y
    W = np.diag(w(x, X, tau))
    XtW = X.T @ W
    return alg.solve( XtW @ X, XtW @ Y)
```

```python
theta(0, X, y)
```

```python
def h(x, X, Y, tau):
    # LWR is non-parametric
    theta_x = theta(x, X, Y, tau)
    return theta_x.T @ np.array([1, x])
```

```python
h(0, X, y)
```

```python
def lwr(xs, X, Y, tau=tau_default):
    return np.array([h(x,X,Y,tau) for x in xs])
```

```python
lwr(np.linspace(-3,3, 20), X, y)
```

```python
def ols(X, Y):
    # minimize least squares using normal equation
    # ordinary giving equal weight to all observation in X
    # assumes constant variance in (homescedasticity)
    return alg.solve(X.T @ X, X.T @ Y)
```

```python
theta_ols = ols(X, y).reshape(d+1,1)
xs = np.linspace(-3, 3, 50)
Xs = np.column_stack([np.ones_like(xs), xs])
print(Xs.shape, theta_ols.shape)
ols_ys = Xs @ theta_ols
plt.plot(xs, ols_ys)
plt.scatter(x, y)
```

```python
fig = plt.figure(figsize=(15,5))
taus = [0.02, 0.1, 1, 3]
for i in range(4):
    axi = fig.add_subplot(1,4, i+1)
    axi.scatter(x, y, s=10, label="datos")
    axi.plot(xs, lwr(xs, X, y, taus[i]), "g-", label=f"LWR tau {taus[i]}")
    axi.plot(xs, ols_ys, "r--", label="OLS")
    axi.set_title(f"tau {taus[i]}")
    axi.legend()
```
