```python
import tensorflow_probability as tfp
```

```python
weather = tfp.distributions.Bernoulli(probs=0.8, name="weather")
```

```python
dataset = weather.sample(10)
print(type(dataset), dataset)
```

```python
# get the probability of the dataset, the probability of each result sample for the known variable model weahter
probas = weather.prob(dataset)
tf.reduce_prod(probas)
```
