```python
import tensorflow as tf
import tensorflow_probability as tfp
```

```python
_W = tfp.distributions.Bernoulli(probs=0.8, name='weather_original')
D = _W.sample(100)
```

```python
#W_fit = tf.Variable(0.5)
p_W = tfp.util.TransformedVariable(0.5, bijector=tfp.bijectors.SoftClip(low=0.0, high=1.0), name='theta_weather_fit')
```

```python
W = tfp.distributions.Bernoulli(probs=p_W, name='weather_fit_distribution')
W.sample(100)
```

```python
LL = lambda : - tf.reduce_sum(W.log_prob(D))
LL()
tfp.math.minimize(loss_fn=LL, optimizer=tf.optimizers.Adam(0.1), num_steps=1000)
```
