```python
import silence_tensorflow.auto
import tensorflow as tf
```

```python
import tensorflow_probability as tfp
```

```python
weather= tfp.distributions.Bernoulli(probs=0.631234, name="weather") 
```

```python
dataset = weather.sample(100)
```

```python
mean = tf.reduce_mean(dataset)
print(f"mean of dataset {mean}")
```

```python
mean = tf.reduce_mean(tf.cast(dataset,float))
print(f"type of dataset is int no float {type(dataset)}")
print(f"mean of dataset {mean}")
```

```python
# we compute the product of the probability W=w[i], this is not common because require know the parameter
joint_probability = tf.reduce_prod(weather.prob(dataset))
print(f"joint probability {joint_probability}")
```
