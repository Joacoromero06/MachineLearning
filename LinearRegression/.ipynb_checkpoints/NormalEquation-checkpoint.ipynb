```python
import numpy as np
```

```python
matrix = [[1,4], [-2, 1], [4,2]]
A = np.array(matrix)

vector = [4, 2, 1]
y = np.array(vector)
```

```python
A_t = A.transpose()
```

```python
A_t @ A 
```

```python
A_t @ y
```

```python
# the normal equation come from the system A_t @ A x = A_t @ y
# x = ( A_t @ A )⁻¹ A_t @ y
```

```python
import numpy.linalg as alg
```

```python
x = alg.inv(A_t @ A) @ (A_t @ y)
print(x)
```
