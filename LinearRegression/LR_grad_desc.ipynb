```python
def h(Xi, Theta):
    # given a input vector sample of n features add a constant value of dimensionality 1
    # at the beggining of the second dimension of Xi
    # given Theta a vector of n+1 parameters
    # computes the dot product or the linear model prediction given Xi and Theta
    Xi = np.hstack((np.array([1]), Xi))
    return np.dot(Xi, Theta)
```

```python
def stochastic_gradiend_descend(X, y, alpha=0.01, n_epoch=100):
    m, n = X.shape
    Theta = np.ones(n+1)
    for _ in range(n_epoch):
        for i in np.random.permutation(m):
            l_e = h(X[i], Theta) - y[i] # lineal error 
            grad = l_e * X[i] * 0.5
            Theta -= alpha * grad
    return Theta
```

```python
print(stochastic_gradiend_descend(X,y))
```

```python
def batch_gradiend_descend(X, y, alpha=0.01, n_iter=1000):
    m, n = X.shape
    Theta = np.ones(n+1)
    ones_matrix = np.ones((m,1))
    X = np.hstack((ones_matrix, X))
    for _ in range(n_iter):
        grad = X.T @ (X@Theta - y) 
        Theta -= (1/m) * alpha * grad
    return Theta

    
batch_gradiend_descend(X, y)
```
